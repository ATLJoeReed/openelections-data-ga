```python
import glob
import os

import pandas as pd
```

```python
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('.', "20131203_*_precinct.csv"))), sort=True)
```

```python
df.info()
```

```python
df.head()
```

```python
columns = {
    'election_day': 'election_day_votes',
    'advance_in_person': 'advanced_votes',
    'absentee_by_mail': 'absentee_by_mail_votes',
    'provisional': 'provisional_votes',
}

df.rename(columns=columns, inplace=True)
```

```python
df.groupby(['party']).count()
```

```python
df.loc[df.party == 'D', 'party'] = 'Democrat'
df.loc[df.party == 'L', 'party'] = 'Libertarian'
df.loc[df.party == 'R', 'party'] = 'Republican'
df.loc[df.party == 'I', 'party'] = 'Independent'
```

```python
df.groupby('office').count()
```

```python
df = df[df['office'].str.startswith('State')]
```

```python
df.groupby('office').count()
```

```python
df.loc[df.office == 'State Representative', 'office'] = 'State House'
df.loc[df.office == 'State Senator', 'office'] = 'State Senate'

df.loc[df.office == 'State Representative, District 100', 'office'] = 'State House'
df.loc[df.office == 'State Representative, District 104', 'office'] = 'State House'
df.loc[df.office == 'State Representative, District 127', 'office'] = 'State House'
df.loc[df.office == 'State Senator, District 14', 'office'] = 'State Senate'
```

```python
df.groupby('office').count()
```

```python
df.head()
```

```python
df.info()
```

```python
df.district = df.district.astype(pd.Int64Dtype())
```

```python
df_clean = df[df['precinct'].notna()]
```

```python
df_clean.head()
```

```python
file_name = '20131203__ga__special__general__runoff.csv'

columns = ['county', 'precinct', 'office', 'district', 'party', 'candidate', 'election_day_votes', 'advanced_votes', 'absentee_by_mail_votes', 'provisional_votes']

df_clean[columns].to_csv(file_name, index=False)
```
