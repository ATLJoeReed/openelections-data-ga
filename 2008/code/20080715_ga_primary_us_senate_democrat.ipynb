```python
import pandas as pd
```

```python
df = pd.read_html('http://sos.ga.gov/elections/election_results/2008_0715/00302.htm')
```

```python
df = df[1]
```

```python
df = df[df.columns[[0,3,4,5,6,7]]]
```

```python
df.columns = ['county', 'Dale Cardwell', 'Vernon Jones', 'Rand Knight', 'Josh Lanier', 'Jim Martin']
```

```python
df.tail(5)
```

```python
df = df.melt(id_vars=['county'])
```

```python
df.columns = ['county', 'candidate', 'votes']
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df.info()
```

```python
df['office'] = 'US Senate'
```

```python
df['district'] = ''
```

```python
df['party'] = 'Democrat'
```

```python
df.groupby(['candidate', 'party'])['votes'].sum()
```
