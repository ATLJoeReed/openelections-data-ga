```python
import pandas as pd
```

```python
df = pd.read_html('http://sos.ga.gov/elections/election_results/2008_1104/012.htm')
```

```python
df = df[1]
```

```python
df.shape
```

```python
df.head(5)
```

```python
df = df[4:]
```

```python
df.head(5)
```

```python
df = df[df.columns[[0,3,4]]]
```

```python
df.head(5)
```

```python
df.columns = ['county', 'H. Doug Everett', 'John H. Monds']
```

```python
df.tail(5)
```

```python
df = df.melt(id_vars=['county'])
```

```python
df.columns = ['county', 'candidate', 'votes']
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df.info()
```

```python
df['office'] = 'Public Service Commission'
```

```python
df['district'] = '1'
```

```python
df.loc[df['candidate'] == 'H. Doug Everett', 'party'] = 'Republican'
df.loc[df['candidate'] == 'John H. Monds', 'party'] = 'Libertarian'
```

```python
df.groupby(['candidate', 'party'])['votes'].sum()
```
