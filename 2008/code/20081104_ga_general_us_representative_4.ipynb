```python
import pandas as pd
```

```python
df = pd.read_html('http://sos.ga.gov/elections/election_results/2008_1104/020.htm')
```

```python
df = df[1]
```

```python
df.shape
```

```python
df.head(5)
```

```python
df = df[4:]
```

```python
df.head(5)
```

```python
df = df[df.columns[[0,3,4,5,6]]]
```

```python
df.head(5)
```

```python
df.columns = ['county', 'Henry C. (Hank) Johnson Jr', 'Faye Coffield', 'Loren Christopher Collins', 'Jacob Perasso']
```

```python
df.tail(5)
```

```python
df = df.melt(id_vars=['county'])
```

```python
df.columns = ['county', 'candidate', 'votes']
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df.info()
```

```python
df['office'] = 'U.S. Representative'
```

```python
df['district'] = '4'
```

```python
df.loc[df['candidate'] == 'Henry C. (Hank) Johnson Jr', 'party'] = 'Democrat'
df.loc[df['candidate'] == 'Faye Coffield', 'party'] = 'Write-In'
df.loc[df['candidate'] == 'Loren Christopher Collins', 'party'] = 'Write-In'
df.loc[df['candidate'] == 'Jacob Perasso', 'party'] = 'Write-In'
```

```python
df.groupby(['candidate', 'office', 'district', 'party'])['votes'].sum()
```
