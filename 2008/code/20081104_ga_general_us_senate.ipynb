```python
import pandas as pd
```

```python
df = pd.read_html('http://sos.ga.gov/elections/election_results/2008_1104/003.htm')
```

```python
df = df[1]
```

```python
df.shape
```

```python
df.head(5)
```

```python
df = df[4:]
```

```python
df.head(5)
```

```python
df = df[df.columns[[0,3,4,5,6,7]]]
```

```python
df.head(5)
```

```python
df.columns = ['county', 'Saxby Chambliss', 'Jim Martin', 'Allen Buckley', 'Eleanor Garcia', 'William Salomone Jr']
```

```python
df.tail(5)
```

```python
df = df.melt(id_vars=['county'])
```

```python
df.columns = ['county', 'candidate', 'votes']
```

```python
df.head()
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df.info()
```

```python
df['office'] = 'US Senate'
```

```python
df['district'] = ''
```

```python
df.loc[df['candidate'] == 'Saxby Chambliss', 'party'] = 'Republican'
df.loc[df['candidate'] == 'Jim Martin', 'party'] = 'Democrat'
df.loc[df['candidate'] == 'Allen Buckley', 'party'] = 'Libertarian'
df.loc[df['candidate'] == 'Eleanor Garcia', 'party'] = 'Write-In'
df.loc[df['candidate'] == 'William Salomone Jr', 'party'] = 'Write-In'
```

```python
df.groupby(['candidate', 'party'])['votes'].sum()
```

```python
df[['county', 'office', 'district', 'party', 'candidate', 'votes']].head()
```
